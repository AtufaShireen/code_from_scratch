```python
def reversell(s, e, p):
    while (s!=e):
        temp = s.next
        s.next = p
        p = s
        s = temp
    return p

def reversegroups(head,k):
    s = head
    e = head
    prev = None
    while (e!=None):
        for _ in range(k):
            e = e.next
        s = reversell(s, e, prev)
    return prev
```
